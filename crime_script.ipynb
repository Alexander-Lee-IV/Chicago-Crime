```python
import pandas as pd
import numpy as np 
from datetime import datetime

crimes_1_4 = pd.read_csv(r'C:\Users\alexa\OneDrive\Desktop\Chicago_crime\Chicago_Crimes_2001_to_2004.csv', on_bad_lines='skip')
crimes_5_7 = pd.read_csv(r'C:\Users\alexa\OneDrive\Desktop\Chicago_crime\Chicago_Crimes_2005_to_2007.csv', on_bad_lines='skip')
crimes_8_11 = pd.read_csv(r'C:\Users\alexa\OneDrive\Desktop\Chicago_crime\Chicago_Crimes_2008_to_2011.csv', on_bad_lines='skip')
crimes_12_17 = pd.read_csv(r'C:\Users\alexa\OneDrive\Desktop\Chicago_crime\Chicago_Crimes_2012_to_2017.csv', on_bad_lines='skip')
```

```python
crimes_1_4 = crimes_1_4.dropna() 
crimes_5_7 = crimes_5_7.dropna()
crimes_8_11 = crimes_8_11.dropna()
crimes_12_17 = crimes_12_17.dropna()
```

```python
dfs = [crimes_1_4, crimes_5_7, crimes_8_11, crimes_12_17]
for df in dfs:
    print(df.shape)
```

```python
crimes_1_4['Year'] = crimes_1_4['Year'].astype('int64')
crimes_5_7['Year'] = crimes_5_7['Year'].astype('int64')
crimes_8_11['Year'] = crimes_8_11['Year'].astype('int64')
crimes_12_17['Year'] = crimes_12_17['Year'].astype('int64')
```

```python
crimes = pd.concat(dfs)
```

```python
crimes.shape
```

```python
crimes[['District', 'Ward', 'Community Area']] = crimes[['District', 'Ward', 'Community Area']].astype(int)
```

```python
crimes
```

Dropping unimportant columns 

```python
crimes = crimes.drop(['Unnamed: 0', 'Ward', 'Community Area', 'X Coordinate', 'Y Coordinate', 'Latitude', 'Longitude'], axis=1)
```

```python
crimes.drop_duplicates(subset=['ID', 'Case Number'], inplace=False)
```

```python
 crimes.dtypes
```

```python
crimes["District"] = crimes["District"].astype(str)
```

```python
crimes = crimes.drop('Month', axis=1)
crimes['Month'] = pd.DatetimeIndex(crimes['Date']).month
```
